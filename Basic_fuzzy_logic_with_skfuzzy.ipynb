```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fz
```

```python
# Defining Universe of discourse(input range)

X = np.arange(0,11,1)
print("Crisp Value:",X)
```

```python
# Defining Set A and set B

A = fz.trimf(X,[0,0,5]) #membership function for set A (e.g, "Low temparature")
B = fz.trimf(X,[5,10,10]) # mebership function for B (e.g, "High temparature")

print("Fuzzy set A:",A)
print("Fuzzy set B:",B)
```

```python
# Fuzzy NOT operation(negation of set A)
A_not = 1 - A
print("Negation:",A_not)
```

```python
#Fuzzy AND operation(Intersaction of set A and Set B using min operator)
A_and_B = np.fmin(A,B)
print("Intersaction:",A_and_B)
```

```python
#Fuzzy OR operation(Union of set A and set B and using the max operator)
A_or_B = np.fmax(A,B)
print("union:",A_or_B)
```

```python
#Visualizing the fuzzy operations

plt.figure(figsize=(10,8))

#ploting set A
plt.subplot(2,2,1)
plt.plot(X,A,label='Set A', color = 'blue')
plt.title("Fuzzy set A")
plt.xlabel("X")
plt.ylabel("Membership Degree")

#ploting set B
plt.subplot(2,2,2)
plt.plot(X,B,label='set B', color='red')
plt.title("Fuzzy set B")
plt.xlabel('X')
plt.ylabel('Membership Degree')

#plotting fuzzy not (negation of set A)
plt.subplot(2,2,3)
plt.plot(X,A_not,label="NOT A",color='green')
plt.title('Fuzzy not A')
plt.xlabel('X')
plt.ylabel('Membership Degree')

#Plotting Fuzzy AND
plt.subplot(2,2,4)
plt.plot(X,A_and_B,label="A and B", color='purple')
plt.title("Fuzzy AND")
plt.xlabel('X')
plt.ylabel("Membership Degree")

#Displaying the result
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Define the universe of discourse
x = np.arange(0, 11, 1)  # Values from 0 to 10

# Define fuzzy sets using triangular membership functions (trimf)
# Syntax: trimf(universe, [a, b, c]) where a < b < c
A = fuzz.trimf(x, [0, 3, 6])  # Triangular membership for set A
B = fuzz.trimf(x, [4, 7, 10])  # Triangular membership for set B

# Plotting the fuzzy sets
plt.figure(figsize=(10, 6))
plt.plot(x, A, 'b', linewidth=1.5, label='Set A')
plt.plot(x, B, 'r', linewidth=1.5, label='Set B')
plt.title('Fuzzy Sets A and B with Triangular Membership Functions')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()

# Fuzzy operations
# Fuzzy Union
A_or_B = np.fmax(A, B)

# Fuzzy Intersection
A_and_B = np.fmin(A, B)

# Fuzzy Complement (not A)
A_not = 1 - A

# Plotting the fuzzy operations
plt.figure(figsize=(10, 6))

# Plot Union
plt.subplot(3, 1, 1)
plt.plot(x, A_or_B, 'g', linewidth=1.5, label='A OR B (Union)')
plt.fill_between(x, 0, A_or_B, color='green', alpha=0.3)
plt.title('Fuzzy Union (A OR B)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()

# Plot Intersection
plt.subplot(3, 1, 2)
plt.plot(x, A_and_B, 'purple', linewidth=1.5, label='A AND B (Intersection)')
plt.fill_between(x, 0, A_and_B, color='purple', alpha=0.3)
plt.title('Fuzzy Intersection (A AND B)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()

# Plot Complement
plt.subplot(3, 1, 3)
plt.plot(x, A_not, 'orange', linewidth=1.5, label='NOT A (Complement)')
plt.fill_between(x, 0, A_not, color='orange', alpha=0.3)
plt.title('Fuzzy Complement (NOT A)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Define the universe of discourse
x = np.arange(0, 11, 1)  # Values from 0 to 10

# Define fuzzy sets using trapezoidal membership functions (trapmf)
# Syntax: trapmf(universe, [a, b, c, d]) where a < b <= c < d
A_trap = fuzz.trapmf(x, [0, 2, 4, 6])  # Trapezoidal membership for set A
B_trap = fuzz.trapmf(x, [3, 5, 7, 9])  # Trapezoidal membership for set B

# Plotting the fuzzy sets
plt.figure(figsize=(10, 6))
plt.plot(x, A_trap, 'b', linewidth=1.5, label='Set A (Trapezoidal)')
plt.plot(x, B_trap, 'r', linewidth=1.5, label='Set B (Trapezoidal)')
plt.title('Fuzzy Sets A and B with Trapezoidal Membership Functions')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()

# Fuzzy operations
# Fuzzy Union (A OR B)
A_or_B = np.fmax(A_trap, B_trap)

# Fuzzy Intersection (A AND B)
A_and_B = np.fmin(A_trap, B_trap)

# Fuzzy Complement (NOT A)
A_not = 1 - A_trap

# Plotting the fuzzy operations
plt.figure(figsize=(10, 8))

# Plot Union
plt.subplot(3, 1, 1)
plt.plot(x, A_or_B, 'g', linewidth=1.5, label='A OR B (Union)')
plt.fill_between(x, 0, A_or_B, color='green', alpha=0.3)
plt.title('Fuzzy Union (A OR B)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)

# Plot Intersection
plt.subplot(3, 1, 2)
plt.plot(x, A_and_B, 'purple', linewidth=1.5, label='A AND B (Intersection)')
plt.fill_between(x, 0, A_and_B, color='purple', alpha=0.3)
plt.title('Fuzzy Intersection (A AND B)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)

# Plot Complement
plt.subplot(3, 1, 3)
plt.plot(x, A_not, 'orange', linewidth=1.5, label='NOT A (Complement)')
plt.fill_between(x, 0, A_not, color='orange', alpha=0.3)
plt.title('Fuzzy Complement (NOT A)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Define the universe of discourse
x = np.arange(0, 11, 1)  # Values from 0 to 10

# Define fuzzy sets using Gaussian membership functions (gaussmf)
# Syntax: gaussmf(universe, mean, sigma) where mean is the center and sigma is the spread
A_gauss = fuzz.gaussmf(x, 3, 1.5)  # Gaussian membership for set A (mean=3, sigma=1.5)
B_gauss = fuzz.gaussmf(x, 7, 1.5)  # Gaussian membership for set B (mean=7, sigma=1.5)

# Plotting the fuzzy sets
plt.figure(figsize=(10, 6))
plt.plot(x, A_gauss, 'b', linewidth=1.5, label='Set A (Gaussian)')
plt.plot(x, B_gauss, 'r', linewidth=1.5, label='Set B (Gaussian)')
plt.title('Fuzzy Sets A and B with Gaussian Membership Functions')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()

# Fuzzy operations
# Fuzzy Union (A OR B)
A_or_B = np.fmax(A_gauss, B_gauss)

# Fuzzy Intersection (A AND B)
A_and_B = np.fmin(A_gauss, B_gauss)

# Fuzzy Complement (NOT A)
A_not = 1 - A_gauss

# Plotting the fuzzy operations
plt.figure(figsize=(10, 8))

# Plot Union
plt.subplot(3, 1, 1)
plt.plot(x, A_or_B, 'g', linewidth=1.5, label='A OR B (Union)')
plt.fill_between(x, 0, A_or_B, color='green', alpha=0.3)
plt.title('Fuzzy Union (A OR B)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)

# Plot Intersection
plt.subplot(3, 1, 2)
plt.plot(x, A_and_B, 'purple', linewidth=1.5, label='A AND B (Intersection)')
plt.fill_between(x, 0, A_and_B, color='purple', alpha=0.3)
plt.title('Fuzzy Intersection (A AND B)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)

# Plot Complement
plt.subplot(3, 1, 3)
plt.plot(x, A_not, 'orange', linewidth=1.5, label='NOT A (Complement)')
plt.fill_between(x, 0, A_not, color='orange', alpha=0.3)
plt.title('Fuzzy Complement (NOT A)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
